```python
import requests
from bs4 import BeautifulSoup
url = "https://www.armurerie-auxerre.com/"
page = requests.get(url)
soup = BeautifulSoup(page.content, 'html.parser')

# print(soup.find_all("a", class_="gem-c-document-list__item-title"))
armes = soup.find_all('div',class_="product")

for mot in armes:
    print(mot.string)
```
